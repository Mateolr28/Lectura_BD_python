#### 1. Ejercicio 1: Datos con Distribución Normal

**Descripción:**

Usaremos un conjunto de datos simulado con una distribución normal para calcular estadísticas descriptivas y visualizar la distribución.

**Código en `Python`**

Recuerden que si alguna de las librerías no les funciona deben usar el comando '!pip install NombreDeLaLibreria'

```python
#Librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
# Generar datos con distribución normal
np.random.seed(42)
datos_normales = np.random.normal(loc=3000, scale=500, size=1000)
df_normal = pd.DataFrame({'Salarios': datos_normales})
```

```python
# Función para análisis descriptivo completo
def analizar_datos(df, columna):
    print(f"\nEstadísticas Descriptivas para {columna}:")
    print(df[columna].describe())
    print(f"\nMedia: {df[columna].mean():.2f}")
    print(f"Mediana: {df[columna].median():.2f}")
    #print(f"Moda: {df[columna].mode()[0]:.2f}")
    print(f"Desviación estándar: {df[columna].std():.2f}")
    print(f"Varianza: {df[columna].var():.2f}")
    print(f"Asimetría: {skew(df[columna]):.2f}")
    print(f"Curtosis: {kurtosis(df[columna]):.2f}")

    # Visualización
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(df[columna], kde=True, bins=30)
    plt.title(f'Distribución de {columna}')
    plt.xlabel(columna)

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[columna])
    plt.title(f'Diagrama de Caja de {columna}')
    plt.show()

```

```python
# Análisis descriptivo de nuestro dataFrame
analizar_datos(df_normal, 'Salarios')
```

**Conclusiones:**

 1. La media y la mediana deben ser similares si la distribución es verdaderamente normal.
 2. La desviación estándar muestra la dispersión alrededor de la media.

#### 2. Ejercicio 2: Datos con Valores Atípicos (Outliers)

**Descripción:**

Generaremos datos con una distribución normal y agregaremos valores atípicos para analizar su impacto.

```python
# Generar datos con valores atípicos
salarios_con_outliers = np.append(np.random.normal(3000, 500, 980), [10000, 12000, 15000, 20000, 25000])
df_outliers = pd.DataFrame({'Salarios': salarios_con_outliers})

# Análisis descriptivo
analizar_datos(df_outliers, 'Salarios')
```

**Conclusiones:**

1. Los valores atípicos afectan significativamente la media y la desviación estándar.
2. El rango intercuartílico (IQR) es útil para identificar estos valores extremos.

#### 3. Ejercicio 3: Datos con Múltiples Modas (Distribución Bimodal)

**Descripción:**

Generaremos un conjunto de datos con dos picos distintos (bimodal) para analizar la presencia de múltiples modas.


```python
# Generar datos bimodales
datos_bimodales = np.concatenate([np.random.normal(2800, 300, 500), np.random.normal(3500, 300, 500)])
df_bimodal = pd.DataFrame({'Salarios': datos_bimodales})

# Análisis descriptivo
analizar_datos(df_bimodal, 'Salarios')
```

**Conclusiones:**

1. La presencia de dos picos indica múltiples modas.
2. La media por sí sola no es representativa en distribuciones multimodales.

#### Conclusión Final

Estos ejercicios permiten comprender cómo varían las métricas descriptivas según la distribución de los datos:

1. En distribuciones normales, la media y la mediana son similares.
2. Los valores atípicos influyen en las medidas de tendencia central y dispersión.
3. Las distribuciones bimodales requieren un análisis más detallado de la moda y la forma de la distribución.
